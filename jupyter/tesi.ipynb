```python
types = csv['typename'].unique()

means = csv.groupby('typename').mean()
variances = csv.groupby('typename').var()
```

```python
print("MEANS\n")
means
```

```python
print("VARIANCES\n")
variances
```

```python
cols = ['time_original','time_columOpt_prealloc','time_columOpt_prealloc_V1','time_fminknap']
titles = ['original','prealloc','prealloc_V1','fminknap']

means[cols].plot.box()
plt.gca().set_xticklabels(titles)
plt.show()

for i, col in enumerate(cols):
    fig, ax = plt.subplots()
    means[col].plot.bar()
    ax.set_title(titles[i])
    ax.set_ylabel('secondi')
    ax.set_xlabel('type')
    plt.show()
```
